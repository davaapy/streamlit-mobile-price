```python
features = ['battery_power', 'm_dep', 'mobile_wt', 'px_height', 'px_width', 'ram', 'talk_time', 'three_g', 'price_range']
x = df[features]
y = df['price_range']
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.45, random_state=90)
y_test.shape
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test,y_test)
print('Akurasi model Regresi Linear', score)
```

```python
print(x)
```

```python
input_data = np.array([[842,0.6,188,20,756,2549,19,0,1]])

prediction = lr.predict(input_data)
print('Estimasi Hasil Mobile Price Classification :',prediction)
print('Nilai Klasifikasi Mobile Price Asli adalah : 2')
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print('R2:', r2_score(y_test,pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, pred)))
```
